```python
# P80_Basic_01_replace, 내용 바꾸기

# replace, 내용 바꾸기
# find, [:], 문자글자찾기,범위 지정
# split 문자열 나누기

var1= "홍길동,"
var2= '25'
var3= " 26"
var4 = var1+var2 + var3

var1 = var1.replace('홍','김')
var1 = var1.replace('김','김씨의김자 ').replace("동",' 선')

var2 = var2.find('5')

print("\n # replace, 내용 바꾸기  ")
print(var1)
print(var2)
print(var1 +", "+ var3)

print("\n # find, [:], 문자글자찾기,범위 지정")
print(var1[0:2])
print(var1[1:])


print('"\n # split 문자열 나누기"')
print(var4.split())
print(var4.split(','))

# uppper, lower , 소문자를대문자로, 대문자를 소문자로 변환
var1_1="hello"
var1_2="HELLO"
var1_3="Hello World"

print(var1_1.upper())
print(var1_2.lower())
print(var1_3.upper().lower())

# len 길이를 출력, 문자열 개수
var1_4 = len(var1_3)+len(var1_2)

print("\n # len 길이를 출력, 문자열 개수")
print(len(var1_3))
print(len(var1_2))
print(len(var1_3)+len(var1_2))
print(var1_4 + 5)

# 공백제거, 공백 제거
kongbai = "   hello world  "
print("\n # 공백제거, 공백 제거")
print(kongbai.rstrip())  # 우측공백 제거
print(kongbai.lstrip())  # 좌측공백 제거
print(kongbai.strip())   # 양측공백 제거
```

```python
# P84_Basic_01_문자열 포맷팅 
# 문자열 str 포맷팅 

var0 = 'I am' + ' 박정태'
print(var0)

print("\n")
var1= "I am {name}".format(name="박정태")
print(var1)

print("\n")
var2= "I am {0}".format("Park jongTae")
# var2= "I am {}".format("Park jongTae")
print(var2)

print("\n")
var3= "I am {0} I am {name}".format("Park jongTae" ,name="정보문화사") 
# I am Park jongTae I am 정보문화사

# var3= "I am {name} I am {0}".format("Park jongTae" ,name="정보문화사")
# I am 정보문화사 I am Park jongTae
print(var3)

print("\n")
# var00 = 'I am age ' + 25  # "+" 연산자는 숫자형과 str 을 더할수 없고 에러남 
# 정확한 방식은 아래와 같음 

age = 25
var00 = "00 I am age " + str(age)   # int -> str 변환 방식 
var01 = "01 I am age {}".format(age)


print(var00)
print(var01)
```

```python
# P86_Basic_01_리스트 자료형 
# 리스트 list 자료형 

var1 = list([1,2,3,4,5])
var01 = [1,2,3,4,5]

print(var1)
print(var01)

print("\n")
# 인덱싱, 슬라이싱하기 
var = [1,2,3,4,5,6,7,8,9]
print("var[0]: " + str(var[0]))
print(var[0])
print(var[1])
print(var[0:3])



print("\n")
# split 문자열 나누기
var = "hello world"
print(var.split())

print("\n")
# join을 이용하여 리스트를 문자열로 만들기 
a = ["1","2","3","4","5"]
b= ".".join(a)
print(a)
print(b)

print("\n")
# 중첩된 리스트
a = [1,2,3,4,[5,6,7]]
print(a)
print(a[3])
print(a[4])
print(a[4][0])
print(a[4][2])
print(a[-1][0])
print(a[-1][1])
print(a[-1][2])
print(a[-1][-1])

print("\n")
# 리스트 연산 + *
# 리스트 연산에서 - / 는 불가능  

var1 = [1,2,3,4,5]
var2 = [6,7,8,9,10]
num = 3
print(var1 * 3)
print(var1 + var2)

print("\n")
# 리스트 수정, 삭제

var = [1,2,3,4,5]
print(var)

var[2] = 10
print(var)

var[3] = [11,22,33]
print(var)

print("\n")
# 리스트 삭제
var = [1,2,3,4,5,6]
del var[1]
print(var)
var2 = [1,2,3,4,5]
var2[0:3]=[]
print(var2)

print("\n")
# 리스트의 다양한 함수

print("\n")
# append() -> 마지막 요소에 추가 
var = [1,2,3,4,5]
var.append(10)
print(var)
var.append(20)
print(var)

print("\n")
# pop() -> 마지막 요소 반환 하면서 리스트에서 마지막 요소 제거 
print(var.pop())
print(var)

print("\n")
# remove() -> 특정 값을 제거할때 사용 
var = [1,2,"a",3,4,5]
print(var.remove("a"))
print(var.remove(3))
print(var)
```
